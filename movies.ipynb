```python
import pandas as pd
df=pd.read_csv('movies.csv')
df.fillna(0,inplace=True)
df
```

```python
#2.	(i). Find no. of movies released under each genre given in the database.
df.groupby(['Major Genre'])['Title'].count()
```

```python
#2.(ii). Find the movies under each genre with 1MDB rating >7 and rotten tomatoes rating > 60.
df1=df[df['Rotten Tomatoes Rating']>60]
df2=df[df['IMDB Rating']>7]
dff=df.loc[(df['Rotten Tomatoes Rating']>60) & (df['IMDB Rating']>7) & (df['Major Genre']!=0)]
#dff
dff.groupby(['Major Genre','Title'])['Rotten Tomatoes Rating','IMDB Rating'].sum()
```

```python
#1.(i). Find out the no of movies released in every month of the year 1995.
import datetime
dfn=df[df['Release Date']!='TBD']
dfn['Release Date']=pd.to_datetime(dfn['Release Date'])

#dfn

dfn[dfn['Release Date'].dt.year==1995].groupby(pd.Grouper(key='Release Date', axis=0, freq='M'))['Title'].count()
```

```python
#1.(ii). Find out the no. of movies released in every year from 1990 to 1998.
dfn[dfn['Release Date'].dt.year.between(1990,1998)].groupby(pd.Grouper(key='Release Date', axis=0, freq='M'))['Title'].count()
```

```python
#3.(i). Find the movies released under each fiction with each director in the ascending order of release dates.
#df1.sort_values(by='Release Date',inplace=True)
df1=df.loc[(df['Director']!=0)&(df['Creative Type']!=0)&(df['Release Date']!='TBD')&(df['Distributor']!=0)&(df['Major Genre']!=0)]
df1['Release Date']=pd.to_datetime(df1['Release Date'])
df1.sort_values(by='Release Date',inplace=True)
df1[['Title','Release Date']]
df1.groupby(['Creative Type','Release Date'])['Title','Director'].sum()
```

```python
#3.(ii). Find movies released under each distributor in the order of genre and director.
df1.groupby(['Distributor','Major Genre','Director'])['Title'].apply(list)
```
