```python
import pandas as pd
df=pd.read_excel('/Adidas_US_Sales_Datasets_1_ (1).xlsx')
del df['Unnamed: 0']
df
```

```python
#1.	List all the retailers with retailer id.
df['Retailer'][df['Retailer ID'].notnull()].unique()
```

```python
#2.	List all the retailers in every region.
pd.set_option('max_colwidth',200)
df.groupby(['Region'])['Retailer'].unique()
```

```python
#3.	List the retailers in every city of a state.
df.groupby(['State','City'])['Retailer'].unique()
```

```python
#4.	List the products sold by the retailer in every city.
df.groupby(['City','Retailer'])['Product'].unique().apply(list)
```

```python
#5.	Find the total sales of every retailer.
df.groupby(['Retailer'])['Total Sales'].sum()
```

```python
#6.	Find the total sales of the retailers in every city along with profit.
df1=df.groupby(['City','Retailer'])[['Total Sales','Operating Profit']].sum()
df1
```

```python
#7.	Find the total sales &profit of each product sold by the retailer.
df.groupby(['Retailer','Product'])[['Total Sales','Operating Profit']].sum()
```

```python
#8.	Find the units sold, total sales &profit of the products sold between the dates 1/1/2020 and 4/15/2020.
df1=df[df['Invoice Date'].between('2020-01-01','2020-04-15')]
df1.groupby(['Product','Units Sold','Invoice Date'])[['Total Sales','Operating Profit']].sum()
```

```python
#9.	Find the no of units sold of each product by each retailer in every city
df.groupby(['Retailer','City','Product'])['Units Sold'].sum()
```

```python
#10.Find the products with different price per unit in different cities with proper information.
df.groupby(['City','Product'])['Price per Unit'].unique().apply(list)
```

```python
#11.Find the retailers who sold the same product with different prices in different cities.
df.groupby(['Product','Price per Unit'])[['Retailer','City']].sum()
```

```python
#12.Find the total sales &profits of all products in every month.
import datetime
df['month']=pd.DatetimeIndex(df['Invoice Date']).month
df.groupby(['month','Product'])[['Total Sales','Operating Profit']].sum()
```

```python
#13.Find the total sales &profit of the products in different sales methods in each city.
df.groupby(['Sales Method','Product'])[['Total Sales','Operating Profit']].sum()
```
