```python
import pandas as pd
df=pd.read_csv('National_Universities_Rankings.csv')
df
```

```python
df[['city','state']]=df['Location'].str.split(',',expand=True)
df
```

```python
#9.	List the universities having tuition fee>15000$ and rank between 120 and  170, inclusive of those  two value
df['In-state']=df['In-state'].str.replace('$','').str.replace(',','').fillna(0).astype(int)
df['Tuition and fees']=df['Tuition and fees'].str.replace('$','').str.replace(',','').fillna(0).astype(int)
df1=df[(df['Tuition and fees']>15000) & (df['Rank'].between(120,170))]
df1[['Name','Tuition and fees','Rank']]
```

```python
#8.	How many universities have marginal fee difference <=$5000 in in-state &out-of state tuition fees
df['marg_diff']=df['Tuition and fees']-df['In-state']
df
df['Name'][df['marg_diff']<=5000]
```

```python
#10.	Find the campuses of universities located in different cities (multiple cities)
df.groupby(['Name'])['city'].apply(list)
```
