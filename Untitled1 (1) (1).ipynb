```python
# 8.Fill the missing values of the data with  average marks of the subject of specific group.
from numpy import NaN as na
cse_data.isnull()
cse_data.fillna({'CN':cse_data['CN'].mean(),'DAA':cse_data['DAA'].mean(),'AFL':cse_data['AFL'].mean(),'OE':cse_data['OE'].mean(),'Professional Elective':cse_data['Professional Elective'].mean()},inplace=True)
it_data.isnull()
it_data.fillna({'CN':it_data['CN'].mean(),'DAA':it_data['DAA'].mean(),'AFL':it_data['AFL'].mean(),'PE':it_data['PE'].mean()},inplace=True)
print(cse_data)
print(it_data)
```
