```python
import pandas as pd
import numpy as np
df=pd.read_csv('/content/National_Universities_Rankings.csv')
df
```

```python
# 1.	Find the universities in which undergraduate students were enrolled.
df1=df[df['Undergrad Enrollment'].notnull()]
df1[['Name']]
```

```python
#2.	Find the universities in which iIn-state students were enrolled.
df2=df[df['In-state'].notnull()]
df2[['Name','In-state']]
```

```python
#3.	Find the universities in which in-state students were not enrolled
df2=df[df['In-state'].isnull()]
df2[['Name','In-state']]
```

```python
#4.	How many universities have both out-of state and in-state students?
df3=df[df['In-state'].notnull()]
df4=df3[df3['Undergrad Enrollment'].notnull()]
df4[['Name']]
```

```python
#5.	List the the cities in which universities were  located.
df[['city','state']]=df['Location'].str.split(",",expand=True)
df[['city']]
```

```python
#6.	List the cities &universities under each state.
df6=df.groupby(['state','city'])['Name'].apply(list)
df6
```

```python
#7.	How many universities in each state have ranking<50? Print them along with ranking.
dff=df[['Name','Rank','state']][df['Rank']<50]
dff.groupby(['state','Name'])['Rank'].apply(list)
```

```python
#8.	How many universities have marginal fee difference <=$5000 in in-state &out-of state tuition fees.
df['In-state']=df['In-state'].str.replace('$','')
df['In-state']=df['In-state'].str.replace(',','')
df['Tuition and fees']=df['Tuition and fees'].str.replace('$','')
df['Tuition and fees']=df['Tuition and fees'].str.replace(',','')
df['In-state'].fillna(0,inplace=True)
df['In-state']=df['In-state'].astype(int)
df['Tuition and fees']=df['Tuition and fees'].astype(int)
df8=df[['Tuition and fees','In-state']]
df['marginal_diff']=df8['Tuition and fees']-df['In-state']
df=df.loc[df['marginal_diff']<=5000]
df['Name'].count()
```

```python
#13.	Find universities with least no of undergraduate students &highest no of graduate students.
min=df[['Name','Undergrad Enrollment']].min()
max=df[['Name','Undergrad Enrollment']].max()
print("least students uni\n",min)
print("\nmost students uni\n",max)
```
